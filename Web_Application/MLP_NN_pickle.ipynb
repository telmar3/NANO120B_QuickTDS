```python
import os
import numpy as np
import pandas as pd
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.neural_network import MLPRegressor

# Set folder path and load data into dictionary
folder_path = '1k files'
data = {}

# Load data into dictionary and replace NaN with column average
for filename in os.listdir(folder_path):
    file_path = os.path.join(folder_path, filename)
    if os.path.isfile(file_path):
        with open(file_path, 'r') as file:
            df1 = pd.read_csv(file_path, header=None, names=['des_flux', 'detrap_en', 'def_conc'], index_col=None, squeeze=True)
            df1 = df1.fillna(df1.mean()) # Replace NaN with column average
            data[filename] = df1

# Combine data into one dataframe
df = pd.concat(data, ignore_index=True)

# Add Gaussian noise to the data
noise_std = 0.1 # set the standard deviation of the noise
noise = np.random.normal(0, noise_std, df['des_flux'].shape)
df_noisy = df['des_flux'] + noise
df_noisy = pd.DataFrame(df_noisy)

# Split data into training and test sets
X_train, X_test, y_train, y_test = train_test_split(df_noisy, df[['def_conc', 'detrap_en']], test_size=0.25, random_state=0)

# Define MLPRegressor model
mlp = MLPRegressor()

# Define parameter grid for GridSearchCV
param_grid = {
    'hidden_layer_sizes': [(50,), (100,), (50, 50), (100, 100)],
    'activation': ['relu', 'tanh', 'logistic'],
    'alpha': [0.0001, 0.001, 0.01, 0.1],
}

# Train the model using GridSearchCV
grid_search = GridSearchCV(mlp, param_grid=param_grid, cv=5, n_jobs=-1)
grid_search.fit(X_train, y_train)

# Print the best hyperparameters and test score
print('Best hyperparameters:', grid_search.best_params_)
print('Test score:', grid_search.score(X_test, y_test))
```

```python
from sklearn.neural_network import MLPRegressor

# Create MLP with best hyperparameters
mlp = MLPRegressor(hidden_layer_sizes=(50, 50), activation='logistic', alpha=0.001, max_iter=1000)

# Train the model
mlp.fit(X_train, y_train)

# Predict using the test set
y_pred = mlp.predict(X_test)

# Calculate the loss on the test set
test_score = mlp.score(X_test, y_test)
print("Test score: ", test_score)
```

```python
import pickle

model_filename = 'MLP_NN_simple.pkl'
pickle.dump(mlp, open(model_filename, 'wb'))
```
